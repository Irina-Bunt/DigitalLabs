```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

```python
plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams['figure.dpi'] = 200
```

```python
# domain
dx = 0.001
L = 1
x = L  * np.arange(-L+dx, L+dx, dx)
n = len(x)
```

```python
# function 1
T = 2 # period
A = 1 # magnitude
f = np.concatenate([A * np.ones(500, dtype=float), A * (-np.ones(500, dtype=float))] * T)
```

```python
# computing series
A0 = np.sum(f * np.ones_like(f)) * dx
fFS = A0 / 2

N = 100
A = np.zeros(N)
B = np.zeros(N)
for k in range(N):
    A[k] = np.sum(f * np.cos(np.pi * (k + 1) * x / L)) * dx
    B[k] = np.sum(f * np.sin(np.pi * (k + 1) * x / L)) * dx
    fFS += A[k] * np.cos((k+1) * np.pi * x * L) + B[k] * np.sin((k+1)*np.pi*x/L)
```

```python
#plotting
plt.title('A[n] & B[n] from n')
plt.xlabel('n')
plt.ylabel('A[n] & B[n]')
plt.plot(np.arange(N), A, label='A[n]', marker='x')
plt.plot(np.arange(N), B, label='B[n]', marker='x')
plt.legend(loc='best')
```

```python
fg, ax = plt.subplots(2, 1, sharex=True)
ax[0].set_title('Fourier Series and Estimating of Error')
ax[0].plot(x, f, 'b', label='x(t)')
ax[0].plot(x, fFS, 'r', label=f'N={N}'.format(N), linewidth=1)
ax[0].set_ylabel('x(t)')
ax[0].legend(loc=1)
ax[1].set_ylabel("x(t) - x^(t)")
ax[1].set_xlabel('T')
ax[1].plot(x, abs(f-fFS), 'r', linewidth=0.5)
```

```python
nquart = int(np.floor(n/4))
f = np.zeros_like(x)
f[nquart:2*nquart] = (4/n) * np.arange(1,nquart+1)
f[2*nquart:3*nquart] = np.ones(nquart) - (4/n)*np.arange(0,nquart)
```

```python
A0 = np.sum(f * np.ones_like(f)) * dx
fFS = A0 / 2

N = 4
A = np.zeros(N)
B = np.zeros(N)
for k in range(N):
    A[k] = np.sum(f * np.cos(np.pi * (k + 1) * x / L)) * dx
    B[k] = np.sum(f * np.sin(np.pi * (k + 1) * x / L)) * dx
    fFS += A[k] * np.cos((k+1) * np.pi * x * L) + B[k] * np.sin((k+1)*np.pi*x/L)
```

```python
#plotting
plt.title('A[n] & B[n] from n')
plt.xlabel('n')
plt.ylabel('A[n] & B[n]')
plt.plot(np.arange(N), A, label='A[n]', marker='x')
plt.plot(np.arange(N), B, label='B[n]', marker='x')
plt.legend(loc='best')
```

```python
fg, ax = plt.subplots(2, 1, sharex=True)
ax[0].set_title('Fourier Series and Estimating of Error')
ax[0].plot(x, f, 'b', label='x(t)')
ax[0].plot(x, fFS, 'r', label=f'N={N}'.format(N), linewidth=1)
ax[0].set_ylabel('x(t)')
ax[0].legend(loc=1)
ax[1].set_ylabel("x(t) - x^(t)")
ax[1].set_xlabel('T')
ax[1].plot(x, abs(f-fFS), 'r', linewidth=0.5)
```
